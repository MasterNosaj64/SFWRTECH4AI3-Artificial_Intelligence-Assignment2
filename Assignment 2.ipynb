Assume the following sample points for input x and output y in the data set where pi=(xi,yi).

p1 = (6, 21); p2 = (10, 40); p3 = (12, 9); p4 = (18, 18); p5 = (20, 60); p6 = (25, 61)

Write a program that answers the following questions. The program must be
in .ipynb format, and runnable in colab. In the program use separate code blocks to
answer each question. The answers must be printed when the program runs.

```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([(6, 21), (10, 40), (12, 9), (18, 18), (20, 60), (25, 61)])

X = points[:, 0]
Y = points[:, 1]
```

1) Write the X and Y matrices needed in the Normal Equation.

```python

X_matrix = np.vstack([np.ones(len(X)), X]).T

padding = max(len(f"{num}") for row in X_matrix for num in row) + 4

header = f"{'X Matrix':<{padding*2}}| {'Y Vector':<{padding}}"
print(header)
print("-" * len(header))

for x_row, y_element in zip(X_matrix, Y):
    x_formatted = " ".join([f"{num:<{padding}.2f}" for num in x_row])
    print(f"{x_formatted}| {y_element:<{padding}.2f}")

print("-" * len(header))
```

2) Using the Normal Equation, calculate the parameters of the best linear
model representing them. (You may use libraries instructions in your
program or write the program from the scratch).

```python
theta = np.linalg.inv(X_matrix.T.dot(X_matrix)).dot(X_matrix.T).dot(Y)

print("Parameters:")
print(theta)
```

3) Estimate the output at these input points.

```python
Y_estimated = X_matrix.dot(theta)
print("Estimated Outputs:")
print(Y_estimated)
```

4) Calculate the MSE cost function for this set of data

```python
MSE = np.mean((Y - Y_estimated)**2)

# Print the MSE
print(f"Mean Squared Error (MSE): {MSE}")
```

5) Plot the ground truth output (output given for each sample point in the data
set) and the estimated outputs versus inputs in a single figure. Use a scatter
plot, and two different colors for the actual and estimated data points.

```python
plt.figure(figsize=(10, 6))

plt.scatter(X, Y, color='blue', label='Actual Outputs')

plt.scatter(X, Y_estimated, color='red', label='Estimated Outputs')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Actual vs Estimated Outputs')
plt.legend()

plt.show()
```
